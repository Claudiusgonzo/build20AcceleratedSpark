```python
files = glob.glob(os.path.join(src, "*.csv"))

files[0:5]
```

```python
def cleanup(f):

    dst = os.path.join(dest, os.path.split(f)[1])
    print(f"Reading: {os.path.split(f)[1]}")
    
    df = pd.read_csv(f, header=None, error_bad_lines=False).replace('', 'N').fillna(78)
    print(f"Writing: {os.path.split(f)[1]}")
    
    df.to_csv(dst, header=None, index=False)
```

```python
def remove_first(f):
    
    dst = os.path.join(dest, os.path.split(f)[1])
    print(f"Reading: {os.path.split(f)[1]}")
    
    with open(f, "r") as fd:
        l=fd.readlines()
        
    print(f"Writing: {os.path.split(f)[1]}")        
    with open(dst, "w") as fd:
        fd.writelines(l[1:])
```

```python
def check(f):
    print(f"Reading... {f}")
    with open(f, "r") as fd:
        lns = fd.readlines()
    empty = []
    for i, l in zip(range(len(lns)), lns):
        if l.find(',,') >= 0 or l.endswith(','): empty += [i]
    if len(empty) > 0: print(f)
    else: print(f"No empty lines: {f}")
```

```python
def clean_schema(f):
    dst = os.path.join(dest, os.path.split(f)[1])
    print(f"Reading: {os.path.split(f)[1]}")
    
    df = pd.read_csv(f, header=None)
    df[0] = 1
    df[5] = 1
    df[6] = 'N'
    df[7] = 145
    df[8] = 226
    df[9] = 2
    df[11] = 0.0
    df[10] = df[12]
    df[16] = df[17]
    df[17] = 0.0
    print(f"Writing: {os.path.split(f)[1]}")        
    df.to_csv(dst, index=False)
```

```python
n_cores = multiprocessing.cpu_count()
pool = Pool(processes=n_cores)
_ = pool.map(remove_first, files)
```
